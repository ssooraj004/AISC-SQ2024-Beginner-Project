```python
for column in db.columns:
    unique_values = db[column].unique()
    print(f"Unique values for '{column}': {unique_values}")
```

```python
columns_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for column_name in columns_with_zero:
    mean_val = db[db[column_name] != 0][column_name].mean()
    db.loc[db[column_name] == 0, column_name] = mean_val
```

```python
db.describe()
```

```python
corr = db.corr()
sns.heatmap(corr, vmin=-1, vmax=1, center=0, annot=True, cmap= 'RdBu')
```

```python
for column in db.columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=db, x=column)
    plt.title('Box Plot of ' + f'{column}')
    plt.xlabel('Values')
    plt.show()
```

```python
for column in db.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=db, x=column, bins=10)
    plt.title(f'Distribution of {column}')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.show()
```
